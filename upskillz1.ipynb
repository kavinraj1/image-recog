```python
import cv2
face_cascade=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
image=cv2.imread("13t8uYk.png")

from matplotlib import pyplot as plt
plt.imshow(image)
```

```python
rgb_image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(rbg_image)
```

```python
gray_image=cv2.cvtColor(rgb_image,cv2.COLOR_RGB2GRAY)
plt.imshow(gray_image)
```

```python
faces=face_cascade.detectMultiScale(gray_image,scaleFactor = 1.3,minNeighbors = 5,minSize = (30,30))
print(faces)
```

```python
image = cv2.imread("13t8uYk.png")
for(x,y,w,h) in faces:
    cv2.rectangle(image,(x,y),(x+w,y+h),(255,0,0),20)
rbg_img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(rbg_img)
```

```python
multi=cv2.imread("multifaces.jpg")

from matplotlib import pyplot as plt
plt.imshow(multi)
```

```python
multi_rgb = cv2.cvtColor(multi,cv2.COLOR_BGR2RGB)
plt.imshow(multi_rgb)
```

```python
multi_gray=cv2.cvtColor(multi_rgb,cv2.COLOR_RGB2GRAY)
plt.imshow(multi_gray)
```

```python
mfaces=face_cascade.detectMultiScale(multi_gray,scaleFactor = 1.3,minNeighbors = 5,minSize = (30,30))
print(mfaces)
```

```python
multi_image = cv2.imread("multifaces.jpg")
for(x,y,w,h) in mfaces:
    cv2.rectangle(multi_image,(x,y),(x+w,y+h),(255,0,0),20)
multi_rgb=cv2.cvtColor(multi_image,cv2.COLOR_BGR2RGB)
plt.imshow(multi_rgb)
```
